```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load Datasets
df_q1 = pd.read_csv('C:/Projetos_AnaliseDados/SQL_Project_Data_Job_Analysis/csv_files/1.Top_Paying_Jobs.csv')
df_q2 = pd.read_csv('C:/Projetos_AnaliseDados/SQL_Project_Data_Job_Analysis/csv_files/2.Top_Paying_Jobs_Skills.csv')
df_q3 = pd.read_csv('C:/Projetos_AnaliseDados/SQL_Project_Data_Job_Analysis/csv_files/3.Top_Demanded_Skills.csv')
df_q4 = pd.read_csv('C:/Projetos_AnaliseDados/SQL_Project_Data_Job_Analysis/csv_files/4.Top_Skills_Salarybased.csv')
df_q5 = pd.read_csv('C:/Projetos_AnaliseDados/SQL_Project_Data_Job_Analysis/csv_files/5.Optimal_skills.csv')
```

```python
print(df_q1.shape)
print(df_q2.shape)
print(df_q3.shape)
print(df_q4.shape)
print(df_q5.shape)
```

```python
df_q1.head(10)
```

```python
# Question: What are the top-paying data analyst jobs?
# Use dark theme style
plt.style.use("dark_background")

# Define a dark-style color palette similar to GitHub's dark theme
colors = sns.color_palette("mako", len(df_q1))

# Sort the DataFrame by salary for better visual clarity
df_q1_sorted = df_q1.sort_values(by="salary_year_avg", ascending=False)

# Plot horizontal bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_q1_sorted,
    x="salary_year_avg",
    y="company_name",
    palette=colors
).set(title="Top 10 Salary per Company")

# Set axis labels
plt.xlabel("Average Yearly Salary")
plt.ylabel("Company")

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
# Question: What skills are required for remote work in the top10-paying data analyst jobs?
# Use dark theme style
plt.style.use("dark_background")

# Define a dark-style color palette similar to GitHub's dark theme
colors = sns.color_palette("mako", len(df_q2))

# Plot horizontal bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_q2,
    x="counting_skills",
    y="skills",
    palette=colors
).set(title="Top 10 Skills for Data Analyst (Based on Top 10 Paying Jobs)")

# Set axis labels
plt.xlabel("Skill Count")
plt.ylabel("Skill")

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
# Question: What are the most in-demand skills for data analysts?

# Use dark theme style
plt.style.use("dark_background")

# Define a dark-style color palette similar to GitHub's dark theme
colors = sns.color_palette("mako", len(df_q3))

# Plot vertical bar chart
plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    data=df_q3,
    x="skills",
    y="demand_count",
    palette=colors
)
barplot.set(title="Most In-demand Skills for Data Analysts")


# Removing Y axe and spine
barplot.yaxis.set_visible(False)
barplot.spines["left"].set_visible(False)
barplot.spines["right"].set_visible(False)
barplot.spines["top"].set_visible(False)



# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Adiciona rtulos de dados no topo de cada coluna
for p in barplot.patches:
    height = p.get_height()
    plt.text(
        p.get_x() + p.get_width() / 2.,
        height + max(df_q3["demand_count"]) * 0.01,  # ligeiramente acima do topo
        f"{height:,.0f}",  # formatao com separador de milhar
        ha="center",
        va="bottom",
        fontsize=9,
        color="white"
    )

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()


```

```python
# Question: What are the top skills based on salary for my role?
# Use dark theme style
plt.style.use("dark_background")

# Define a dark-style color palette similar to GitHub's dark theme
colors = sns.color_palette("mako", len(df_q4))

# Plot horizontal bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_q4,
    x="avg_salary",
    y="skills",
    palette=colors
).set(title="Top skills based on salary for Data Analysts")


# Set axis labels
plt.xlabel("Average Salary")
plt.ylabel("Skills")

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()


```

```python
# Question: What are the most optimal skills to learn (aka it's in high demand and a high-paying skill)?

plt.style.use("dark_background")

plt.figure(figsize=(12, 6))

scatter = sns.scatterplot(
    data=df_q5,
    x="demand_count",
    y="avg_salary",
    hue="skills",
    palette="mako",
    s=150,  # tamanho dos pontos
    legend=False  # oculta legenda automtica das skills se tiver muitas
)

plt.title("Optimal Skills to Learn as a Data Analyst", fontsize=14)
plt.xlabel("Demand Count")
plt.ylabel("Average Salary")

# Anota cada ponto com o nome da skill
for i in range(df_q5.shape[0]):
    plt.text(
        df_q5["demand_count"][i] + 3,  # ajuste de posio horizontal
        df_q5["avg_salary"][i],
        df_q5["skills"][i],
        fontsize=9,
        color="white"
    )

plt.tight_layout()
plt.show()
```
